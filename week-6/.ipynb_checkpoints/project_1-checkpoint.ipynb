```python
import tkinter as tk
from tkinter import messagebox

def ibejulekki1(name):
    window = tk.Toplevel(root)
    window.title("Package Details")
    window.geometry("500x200")
 
    label_1 = tk.Label(window, text=f"Welcome {name}.\n")
    label_1.pack()
    label_2 = tk.Label(window, text="Your package is equal to or greater than 10kg and being delivered to the Ibeju-Lekki community.")
    label_2.pack()
    label_3 = tk.Label(window, text="You will be charged N5000 for delivery.")
    label_3.pack()

    root.mainloop()


def ibejulekki2(name):
    window = tk.Toplevel(root)
    window.title("Package Details")
    window.geometry("500x200")

    label_1 = tk.Label(window, text=f"Welcome {name}.\n")
    label_1.pack()
    label_2 = tk.Label(window, text="Your package is less than 10kg and being delivered to the Ibeju-Lekki community.")
    label_2.pack()
    label_3 = tk.Label(window, text="You will be charged N3500 for delivery.")
    label_3.pack()

    root.mainloop()


def epe1(name):
    window = tk.Toplevel(root)
    window.title("Package Details")
    window.geometry("500x200")

    label_1 = tk.Label(window, text=f"Welcome {name}.\n")
    label_1.pack()
    label_2 = tk.Label(window, text="Your package is equal to or greater than 10kg and being delivered to Epe.")
    label_2.pack()
    label_3 = tk.Label(window, text="You will be charged N10000 for delivery.")
    label_3.pack()

    root.mainloop()


def epe2(name):
    window = tk.Toplevel(root)
    window.title("Package Details")
    window.geometry("500x200")

    label_1 = tk.Label(window, text=f"Welcome {name}.\n")
    label_1.pack()
    label_2 = tk.Label(window, text="Your package is less than 10kg and being delivered to the Epe.")
    label_2.pack()
    label_3 = tk.Label(window, text="You will be charged N5000 for delivery.")
    label_3.pack()

    root.mainloop()


def submit():
    name = name_entry.get()
    weight = int(weight_entry.get())
    location = location_entry.get().lower()

    if weight >= 10 and location == "ibeju-Lekki":
        ibejulekki1(name)
    elif weight < 10 and location == "ibeju-Lekki":
        ibejulekki2(name)
    elif weight >= 10 and location == "epe":
        epe1(name)
    elif weight < 10 and location == "epe":
        epe2(name)
    else:
        messagebox.showerror("Error", "We do not deliver to this location")


root = tk.Tk()
root.title("Delivery Form")
root.geometry("500x200")

name_label = tk.Label(root, text="Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

weight_label = tk.Label(root, text="Weight of the package(s) (kg):")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

location_label = tk.Label(root, text="Delivery location:")
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
